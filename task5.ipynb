```python
import numpy as np
from networkx import nx
import matplotlib.pyplot as plt
```

```python
# Function for graph generation
def generate_graph(v=100, e=200):
    graph = nx.gnm_random_graph(v, e)
    return graph
```

```python
# Function for getting an adjacency matrix
def adjacency_matrix(graph):
    adj_matrix = [[0 for _ in range(len(graph))] for _ in range(len(graph))]
    for edge in graph.edges:
        adj_matrix[edge[0]][edge[1]] = adj_matrix[edge[1]][edge[0]]  = 1
    return adj_matrix
```

###  Depth-first search 

```python
def dfs(temp, v, visited): 
        # Mark the current vertex as visited 
        visited[v] = True
  
        # Store the vertex to list 
        temp.append(v) 
  
        # Repeat for all vertices adjacent to this vertex
        for i in adj_list[v]: 
            if visited[i] == False: 
                  
                # Update the list 
                temp = dfs(temp, i, visited) 
        return temp
```

### Breadth-first search 

```python
# Function for find the shortest path between two vertices using BFS
def bfs_shortest_path(start, goal):
    
    # save all visited vertices
    visited = []
    # create queue for all vertices which should be visited
    queue = [[start]]
 
    while queue:
        # get the first path from the queue
        path = queue.pop(0)

        node = path[-1]
        if node not in visited:
            neighbours = adj_list[node]
            # go through all neighbour nodes, construct a new path and push it into the queue
            for neighbour in neighbours:
                new_path = list(path)
                new_path.append(neighbour)
                queue.append(new_path)

                if neighbour == goal:
                    return new_path
 
            # mark node as explored
            visited.append(node)
    return 'No path'
```
